# Network graph plotting

---
First, we load the package modules.

```python
import spaceplot_DMtools as spdmt # tools to load/manipulate distance matrices
import spaceplot_plotter as spp # plotting tools
```

## Load data

Load DM and labels:

```python
labels,DM = spdmt.read_DM( 'example_DM.csv')
```

## Plot and save distance matrix

```python
help(spp.DM_to_NetworkGraph_plot)
spp.DM_to_NetworkGraph_plot( labels, DM )

from IPython.display import Image # to display Graphviz plot in notebook
Image(filename='NetworkGraph_plot.png') 
```
