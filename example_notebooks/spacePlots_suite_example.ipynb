# spacePlots suite example

Here is an example of the batch spacePlots function to explore a 2D distance matrix. Included are:
- Distance matrix plot
- Network graph plot
- Axes relationship plot

All that is needed, as input, is the distance matrix CSV.

---
First, we load the package modules.

```python
import spaceplot_suite as sps
```

## Plot suite

```python
sps.spacePlot_suite('example_DM.csv', 
                    figlist = ['DM','Network'], 
                    axes_r = [-.3,['Trustworthiness','Dominance']])
```
